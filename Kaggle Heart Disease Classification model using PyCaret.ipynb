```python
import pandas as pd
from pycaret.classification import *
```

## Read data into pandas DataFrame

```python
df = pd.read_csv('heart.csv')
```

```python
cat_features = ['sex','cp','fbs','restecg','exang','slope','ca','thal']
```

## Train and Evaluate Model

```python
experiment = setup(df, target='target', categorical_features=cat_features)
```

```python
best_model = compare_models()
```

## Test Model

```python
predict_model(best_model, df.tail())
```

```python
ridge = create_model('ridge')
```

```python
predict_model(ridge, df.tail())
```

## Save Model

```python
save_model(best_model, model_name='logistic-regression-heart-disease')
```

## Load saved model and test on new data

```python
model = load_model('logistic-regression-heart-disease')
```

```python
model.predict(df.tail())
```
